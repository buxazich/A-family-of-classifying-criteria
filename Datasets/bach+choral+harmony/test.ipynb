```python
import pandas as pd
from pathlib import Path

# папка с .data файлами
DATA_DIR = Path("/Users/user/Downloads/led+display+domain")

all_dfs = []

for file in DATA_DIR.glob("*.data"):

    try:
        # читаем csv-like файл
        df = pd.read_csv(file, header=None)

        # имя датасета
        dataset_name = file.stem

        # допустим target = последний столбец
        target_col = df.columns[-1]

        # число классов
        n_classes = df[target_col].nunique()

        # число фичей
        n_features = len(df.columns) - 1

        # фильтр
        if n_classes > n_features:

            # переименуем target
            df = df.rename(columns={target_col: "target"})

            # добавим имя датасета
            df["dataset_name"] = dataset_name

            all_dfs.append(df)

            print(
                f"added: {dataset_name} "
                f"(classes={n_classes}, features={n_features})"
            )

    except Exception as e:
        print("skip:", file.name, e)

# ==================================
# объединение
# ==================================

merged = pd.concat(all_dfs, ignore_index=True, sort=False)

merged.to_csv("yeast.csv", index=False)

print("\nSaved -> merged_dataset.csv")
print("shape:", merged.shape)
```
